```python
# code to datasets which shows only new cases daily
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# upload csv
csv = 'full_data.csv'
df_hist = pd.read_csv(csv,sep=';|,')

#order data to ensure chronology
df_hist.timestamp = pd.to_datetime(df_hist.timestamp)
df_hist = df_hist.sort_values(by=['country','timestamp'])

#create a column of acumulated cases
df_hist['acumulated'] = 0

#rename the 'count' column to "new_cases"
df_hist = df_hist.rename(columns={'count':'new_cases'})

#reset index
df_hist = df_hist.reset_index()

#iterate to sum acumulated cases
for index, row in df_hist.iterrows():

    if index != 0:
        if df_hist.iloc[index-1]['country']==df_hist.iloc[index]['country']:
            df_hist.at[index,'acumulated'] = df_hist.iloc[index-1]['acumulated']+df_hist.iloc[index-1]['new_cases']
        else:
            print("")
    
# print dos nomes dos grupos
print(df_hist.groupby('country').groups.keys())

#plot specific countries
try:
    fig3, ax3 = plt.subplots(figsize=(8,6))
    df_hist.groupby('country').get_group('Brazil').plot(x='acumulated',y='new_cases',ax=ax3,label="Brazil", loglog=True)
    df_hist.groupby('country').get_group('France').plot(x='acumulated',y='new_cases',ax=ax3, label="France", loglog=True)
    df_hist.groupby('country').get_group('Italy').plot(x='acumulated',y='new_cases',ax=ax3, label="Italy", loglog=True)
    df_hist.groupby('country').get_group('China').plot(x='acumulated',y='new_cases',ax=ax3, label="China", loglog=True)
    df_hist.groupby('country').get_group('United States').plot(x='acumulated',y='new_cases',ax=ax3, label="US", loglog=True)
    df_hist.groupby('country').get_group('South Korea').plot(x='acumulated',y='new_cases',ax=ax3, label="S Korea", loglog=True)
except:
    pass
    
#plot normal chart
fig, ax = plt.subplots(figsize=(8,6))
df_hist.groupby('country').plot(x='acumulated',y='new_cases',ax=ax)
ax.legend(df_hist.groupby('country').groups.keys())

#plot log chart
fig2, ax2 = plt.subplots(figsize=(8,6))
df_hist.groupby('country').plot(x='acumulated',y='new_cases',ax=ax2, loglog=True)
ax2.legend(df_hist.groupby('country').groups.keys())
```

```python
# code to datasets which shows only acumulated cases
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# upload csv
csv = 'covid_br_state.csv'
df_hist = pd.read_csv(csv,sep=';|,')

#order data to ensure chronology
df_hist.timestamp = pd.to_datetime(df_hist.timestamp)
df_hist = df_hist.sort_values(by=['country','timestamp'])

#create a column of acumulated cases
df_hist['new_cases'] = 0

#rename the 'count' column to "new_cases"
df_hist = df_hist.rename(columns={'count':'acumulated'})

#reset index
df_hist = df_hist.reset_index()

#iterate to sum acumulated cases
for index, row in df_hist.iterrows():

    if index != 0:
        if df_hist.iloc[index-1]['country']==df_hist.iloc[index]['country']:
            df_hist.at[index,'new_cases'] = df_hist.iloc[index]['acumulated']-df_hist.iloc[index-1]['acumulated']
        else:
            print("")
    
# print dos nomes dos grupos
print(df_hist.groupby('country').groups.keys())

#plot specific counetries
try:
    fig3, ax3 = plt.subplots(figsize=(8,6))
    df_hist.groupby('country').get_group('SP').plot(x='acumulated',y='new_cases',ax=ax3, label="SP", loglog=True)
    df_hist.groupby('country').get_group('RJ').plot(x='acumulated',y='new_cases',ax=ax3, label="RJ", loglog=True)
    df_hist.groupby('country').get_group('DF').plot(x='acumulated',y='new_cases',ax=ax3, label="DF", loglog=True)
except:
    pass
    
#plot normal chart
fig, ax = plt.subplots(figsize=(8,6))
df_hist.groupby('country').plot(x='acumulated',y='new_cases',ax=ax)
ax.legend(df_hist.groupby('country').groups.keys())

#plot log chart
fig2, ax2 = plt.subplots(figsize=(8,6))
df_hist.groupby('country').plot(x='acumulated',y='new_cases',ax=ax2, loglog=True)
ax2.legend(df_hist.groupby('country').groups.keys())
```
